```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lv_solution = np.loadtxt('lv_sol.dat', unpack=True)
plt.plot(lv_solution[1], lv_solution[2], label='orbit one')
plt.plot(lv_solution[3], lv_solution[4], label='orbit two')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

```python
plt.plot(lv_solution[0], lv_solution[9], label='Energy')
plt.xlabel('time')
plt.ylabel('Energy')
plt.legend()
plt.show()
```
