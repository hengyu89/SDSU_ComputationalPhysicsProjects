```python
# Importing numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
file_name = 'density_results.dat'

all_data = np.loadtxt(file_name,skiprows=1)

all_data = all_data.transpose()

position_x = all_data[0]
initial_probability_density = all_data[1]
second_probability_density = all_data[2]
third_probability_density = all_data[3]
fourth_probability_density = all_data[4]
```

```python
file_name2 = 'time_results.dat'

all_data = np.loadtxt(file_name2, skiprows=1)

all_data = all_data.transpose()

time = all_data[0]
normalization = all_data[1]
position = all_data[2]
analytic_position = all_data[3]
numerical_width = all_data[4]
analytic_width = all_data[5]
```

```python
# I left the potential = 0 (get rid of advance part) to show the graph is same as example on README.md.
plt.plot(position_x,initial_probability_density, label = 't = 0')
plt.plot(position_x,second_probability_density, label = 't = 1.65')
plt.plot(position_x,third_probability_density, label = 't = 3.35')
plt.plot(position_x,fourth_probability_density, label = 't = 5')

plt.xlabel('x')
plt.ylabel('p(x)')

plt.legend()
plt.show()
```

```python
# Same conditions as the graph above
plt.plot(time, numerical_width, label = 'numerical result')
plt.plot(time, analytic_width, label = 'analytic result')

plt.xlabel('t')
plt.ylabel('sigma')

plt.legend()
plt.show()
```

```python
plt.plot(time, normalization, label = 'normalization')

plt.xlabel('t')
plt.ylabel('normalization')

plt.legend()
plt.show()
```

```python
plt.plot(time, position, label = 'position')

plt.xlabel('t')
plt.ylabel('position')

plt.legend()
plt.show()
```

```python
plt.plot(time, position, label = 'position')
plt.plot(time, analytic_position, label = 'analytic_position')

plt.xlabel('t')
plt.ylabel('position')

plt.legend()
plt.show()
# Two curves are overlapping perfectly
```
